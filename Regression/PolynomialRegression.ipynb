```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
```

```python
for line in open('data/data_poly.csv'):
    x, y = line.split(',')
    x = float(x)
    X.append([1, x, x*x])
    Y.append(float(y))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X[:,1], Y)
plt.show()
```

```python
# calculate weights
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
```

```python
Yhat = np.dot(X, w)
```

```python
plt.scatter(X[:,1], Y)
plt.plot(sorted(X[:,1]), sorted(Yhat))
plt.show()
```

```python
SSres = sum((Y-Yhat)**2)
SStot = sum((Y-Y.mean())**2)
R2 = 1 - SSres/SStot
print('R-squared is: ', R2)
```
